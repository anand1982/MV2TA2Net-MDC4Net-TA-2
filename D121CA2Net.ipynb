```python
# ==============================================================
# Imports
# ==============================================================
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import DenseNet121
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
from itertools import cycle
import seaborn as sns

# ==============================================================
# Configuration
# ==============================================================
data_dir = "/kaggle/input/ccmt-2008-mnit/New folder"
img_size = (224, 224)
batch_size = 32
seed = 42
epochs = 50
lr = 1e-4
model_path = "efficientnetb0_channel_attention.h5"

# ==============================================================
# Dataset
# ==============================================================
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="training",
    seed=seed,
    image_size=img_size,
    batch_size=batch_size,
    label_mode="categorical"
)

test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="validation",
    seed=seed,
    image_size=img_size,
    batch_size=batch_size,
    label_mode="categorical"
)

class_names = train_ds.class_names
num_classes = len(class_names)
print("Classes:", class_names)

AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.cache().prefetch(AUTOTUNE)
test_ds = test_ds.cache().prefetch(AUTOTUNE)

# ==============================================================
# Channel Attention (SE Block)
# ==============================================================
def channel_attention(x, ratio=8):
    ch = x.shape[-1]

    avg = layers.GlobalAveragePooling2D()(x)
    avg = layers.Dense(ch // ratio, activation="relu")(avg)
    avg = layers.Dense(ch, activation="sigmoid")(avg)

    maxp = layers.GlobalMaxPooling2D()(x)
    maxp = layers.Dense(ch // ratio, activation="relu")(maxp)
    maxp = layers.Dense(ch, activation="sigmoid")(maxp)

    scale = layers.Add()([avg, maxp])
    scale = layers.Reshape((1, 1, ch))(scale)

    return layers.Multiply()([x, scale])

# ==============================================================
# Model: EfficientNetB0 + Channel Attention
# ==============================================================
def build_model():
    inputs = layers.Input(shape=img_size + (3,))
    x = layers.Rescaling(1./127.5, offset=-1)(inputs)

    base = DenseNet121(
        include_top=False,
        weights="imagenet",
        input_tensor=x
    )
    base.trainable = True

    x = base.output
    x = channel_attention(x)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(256, activation="relu")(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)

    return models.Model(inputs, outputs, name="EfficientNetB0_ChannelAttention")

model = build_model()
model.summary()

model.compile(
    optimizer=tf.keras.optimizers.Adam(lr),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

# ==============================================================
# Training
# ==============================================================
model.fit(
    train_ds,
    validation_data=test_ds,
    epochs=epochs,
    callbacks=[
        tf.keras.callbacks.ModelCheckpoint(
            model_path,
            monitor="val_accuracy",
            save_best_only=True,
            verbose=1
        )
    ]
)

# ==============================================================
# Evaluation
# ==============================================================
images_all, y_true, y_pred, y_prob = [], [], [], []

for imgs, labels in test_ds:
    preds = model.predict(imgs, verbose=0)
    images_all.extend(imgs.numpy())
    y_true.extend(np.argmax(labels.numpy(), axis=1))
    y_pred.extend(np.argmax(preds, axis=1))
    y_prob.extend(np.max(preds, axis=1))

images_all = np.array(images_all)
y_true = np.array(y_true)
y_pred = np.array(y_pred)
y_prob = np.array(y_prob)

print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

# ==============================================================
# Confusion Matrix
# ==============================================================
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Greens",
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()

# ==============================================================
# ROC Curve
# ==============================================================
y_true_bin = label_binarize(y_true, classes=range(num_classes))
y_score = model.predict(test_ds)

fpr, tpr, roc_auc = {}, {}, {}
for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure(figsize=(8,6))
for i, color in zip(range(num_classes), cycle(plt.cm.tab10.colors)):
    plt.plot(fpr[i], tpr[i], color=color,
             label=f"{class_names[i]} (AUC={roc_auc[i]:.2f})")

plt.plot([0,1],[0,1],'k--')
plt.legend()
plt.title("ROC Curve")
plt.show()

# ==============================================================
# Correct & Misclassified Samples
# ==============================================================
def show_samples(indices, title):
    plt.figure(figsize=(18,4))
    for i, idx in enumerate(indices[:5]):
        img = (images_all[idx] + 1) / 2
        plt.subplot(1,5,i+1)
        plt.imshow(img)
        plt.axis("off")
        plt.title(
            f"T:{class_names[y_true[idx]]}\n"
            f"P:{class_names[y_pred[idx]]}\n"
            f"C:{y_prob[idx]*100:.1f}%"
        )
    plt.suptitle(title)
    plt.show()

correct_idx = np.where(y_true == y_pred)[0]
wrong_idx = np.where(y_true != y_pred)[0]

show_samples(correct_idx, "Correctly Classified Samples")
show_samples(wrong_idx, "Misclassified Samples")

# ==============================================================
# Uncertainty + Grad-CAM
# ==============================================================
def get_last_conv_layer(model):
    for layer in reversed(model.layers):
        if isinstance(layer, layers.Conv2D):
            return layer.name

def gradcam(model, img):
    layer_name = get_last_conv_layer(model)
    grad_model = tf.keras.Model(
        model.inputs,
        [model.get_layer(layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_out, preds = grad_model(img)
        class_idx = tf.argmax(preds[0])
        loss = preds[:, class_idx]

    grads = tape.gradient(loss, conv_out)
    pooled = tf.reduce_mean(grads, axis=(0,1,2))
    heatmap = tf.reduce_sum(conv_out[0] * pooled, axis=-1)
    heatmap = tf.maximum(heatmap, 0)
    heatmap /= tf.reduce_max(heatmap) + 1e-8
    return heatmap.numpy()

entropy = -np.sum(y_score * np.log(y_score + 1e-8), axis=1)
top_uncertain = np.argsort(entropy)[-5:]

plt.figure(figsize=(18,4))
for i, idx in enumerate(top_uncertain):
    img = (images_all[idx] + 1) / 2
    cam = gradcam(model, tf.expand_dims(images_all[idx],0))
    cam = tf.image.resize(cam[...,None], img.shape[:2]).numpy().squeeze()

    plt.subplot(1,5,i+1)
    plt.imshow(img)
    plt.imshow(cam, cmap="jet", alpha=0.5)
    plt.axis("off")
    plt.title(
        f"{class_names[y_pred[idx]]}\nEntropy:{entropy[idx]:.3f}"
    )

plt.suptitle("Uncertainty-Aware Heatmaps (EfficientNetB0)")
plt.show()

# ==============================================================
# Save Model
# ==============================================================
model.save("final_" + model_path)
print("Model saved successfully.")
```
