```python
import tensorflow as tf
from tensorflow.keras import regularizers

# Define parameters
img_height, img_width = 224, 224
batch_size = 32
validation_split = 0.2  # 20% of the data for validation

# Load the training dataset
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    '/kaggle/input/ccmt-2008-mnit/New folder',
    validation_split=validation_split,
    subset="training",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size
)

# Load the validation dataset
test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    '/kaggle/input/ccmt-2008-mnit/New folder',
    validation_split=validation_split,
    subset="validation",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size
)

# Extract class names
class_names = train_ds.class_names
print("Class names:", class_names)

# Normalize the data
normalization_layer = tf.keras.layers.Rescaling(1./255)
train_ds = train_ds.map(lambda x, y: (normalization_layer(x), y))
test_ds = test_ds.map(lambda x, y: (normalization_layer(x), y))

# Add prefetching to improve performance
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.prefetch(buffer_size=AUTOTUNE)
test_ds = test_ds.prefetch(buffer_size=AUTOTUNE)


from tensorflow.keras.applications import MobileNetV2, DenseNet121
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Concatenate, Input
from tensorflow.keras.models import Model

# Input layer (shared for both MobileNet and ResNet)
input_layer = Input(shape=(img_height, img_width, 3))

# MobileNetV2 branch
mobilenet = MobileNetV2(
    input_shape=(img_height, img_width, 3),
    include_top=False,
    weights='imagenet'
)(input_layer)

mobilenet.trainable = True 
# ResNet50 branch
densenet = DenseNet121(
    input_shape=(img_height, img_width, 3),
    include_top=False,
    weights='imagenet'
)(input_layer)

densenet.trainable = True

# Global Average Pooling
mobilenet_output = GlobalAveragePooling2D()(mobilenet)
densenet_output = GlobalAveragePooling2D()(densenet)

# Concatenate the outputs
concatenated = Concatenate()([mobilenet_output, densenet_output])

# Final classification layer
output = Dense(len(class_names), activation='softmax', kernel_regularizer=regularizers.l2(0.5))(concatenated)

# Create the model
model = Model(inputs=input_layer, outputs=output)

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    train_ds,
    validation_data=test_ds,
    epochs=50
)



# Evaluate the model
test_loss, test_acc = model.evaluate(test_ds)
print(f'Test accuracy: {test_acc:.2f}')

# Evaluate the model
train_loss, train_acc = model.evaluate(train_ds)
print(f'Train accuracy: {train_acc:.2f}')


# Save the trained model
model.save('plant_village_mobilenet_resnet.h5')


import matplotlib.pyplot as plt

# Plot training & validation accuracy values
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Validation'])

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Validation'])

plt.show()


model.summary()


import cv2
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report

# Predict on the test dataset
y_pred = []
y_true = []

# Collect true labels and predictions
for images, labels in test_ds:
    y_true.extend(labels.numpy())
    predictions = model.predict(images)
    y_pred.extend(np.argmax(predictions, axis=1))

# Ensure the number of unique labels matches
print(f"Number of unique true labels: {len(np.unique(y_true))}")
print(f"Number of unique predicted labels: {len(np.unique(y_pred))}")

# If there's a mismatch, specify the labels manually in the classification report
unique_labels = np.unique(y_true)
class_names = [class_names[i] for i in unique_labels]

# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred, labels=unique_labels)
print("Confusion Matrix:")
print(cm)

# Classification report for precision, recall, and F1-score
report = classification_report(y_true, y_pred, target_names=class_names, labels=unique_labels)
print("Classification Report:")
print(report)


from tensorflow.keras.preprocessing import image

# Load and preprocess the image
img_path = '/kaggle/input/sugarcane-img-dis-9026/sugarcane-disease_9026img/test/Grassy shoot/Amit_000_P_010_-_Copy.jpg'
img = image.load_img(img_path, target_size=(img_height, img_width))
img_array = image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0)  # Create a batch

# Normalize the image
img_array = normalization_layer(img_array)

# Predict the class
predictions = model.predict(img_array)
predicted_class = class_names[np.argmax(predictions)]
print(f'The image is predicted to be: {predicted_class}')

# Define Anchor Boxes (x, y, width, height) on the test image
anchor_boxes = [
    (50, 100, 100, 150),  # Anchor box 1
    (150, 200, 120, 160),  # Anchor box 2
    (250, 300, 140, 180)   # Anchor box 3
]

def multi_disease_detection(model, img, anchor_boxes):
    detected_boxes = []
    
    # Iterate over each anchor box
    for (x, y, w, h) in anchor_boxes:
        region = img[y:y+h, x:x+w]  # Crop the region of the image
        region_resized = cv2.resize(region, (224, 224)) / 255.0
        region_expanded = np.expand_dims(region_resized, axis=0)
        
        # Predict on the cropped region
        prediction = model.predict(region_expanded)
        predicted_class = np.argmax(prediction)
        
        # Append if disease detected
        detected_boxes.append((x, y, w, h, predicted_class))
    
    return detected_boxes

# Test on an input image
img = cv2.imread('/kaggle/input/sugarcane-img-dis-9026/sugarcane-disease_9026img/test/Grassy shoot/Grassy-Shoot-Management-in-Sugarcane-1.jpg')
detected_diseases = multi_disease_detection(model, img, anchor_boxes)

# Display detected regions and their predicted classes
for (x, y, w, h, disease_class) in detected_diseases:
    print(f"Disease detected in box [{x}, {y}, {w}, {h}] with predicted class: {disease_class}")

    
    
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image

# Define Anchor Boxes (x, y, width, height)
anchor_boxes = [
    (50, 100, 100, 150),  # Anchor box 1
    (150, 200, 120, 160),  # Anchor box 2
    (250, 300, 140, 180)   # Anchor box 3
]

# Dictionary for class labels


class_labels = ['Banded Chlorosis', 'Brown Spot', 'BrownRust', 'Dried Leaves', 'Grassy shoot', 'Healthy Leaves', 'Pokkah Boeng', 'Sett Rot', 'Viral Disease', 'Yellow Leaf', 'smut']

# Function to perform disease detection and display results on the image
def multi_disease_detection(model, img_path, anchor_boxes, class_labels):
    # Load the image
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for display
    
    # Iterate over each anchor box
    for (x, y, w, h) in anchor_boxes:
        # Crop the region of interest
        region = img[y:y+h, x:x+w]
        region_resized = cv2.resize(region, (224, 224)) / 255.0
        region_expanded = np.expand_dims(region_resized, axis=0)
        
        # Predict the disease class in the cropped region
        prediction = model.predict(region_expanded)
        predicted_class = np.argmax(prediction)
        predicted_label = class_labels[predicted_class]
        
        # Draw the anchor box on the image
        color = (255, 0, 0)  # Red color for the anchor box
        cv2.rectangle(img_rgb, (x, y), (x + w, y + h), color, 2)
        
        # Put the predicted label above the box
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(img_rgb, predicted_label, (x, y - 10), font, 0.5, color, 2)
    
    # Display the result using Matplotlib
    plt.figure(figsize=(8, 8))
    plt.imshow(img_rgb)
    plt.title('Detected Diseases with Anchor Boxes')
    plt.axis('off')  # Hide the axes
    plt.show()

# Path to the input test image
test_img_path = '/kaggle/input/sugarcane-img89/imag89.jpg'

# Call the function with the trained model, test image path, and anchor boxes
multi_disease_detection(model, test_img_path, anchor_boxes, class_labels)

# Define Anchor Boxes (x, y, width, height) on the test image
anchor_boxes = [
    (50, 100, 100, 150),  # Anchor box 1
    (150, 200, 120, 160),  # Anchor box 2
    (250, 300, 140, 180)   # Anchor box 3
]

def multi_disease_detection(model, img, anchor_boxes):
    detected_boxes = []
    
    # Iterate over each anchor box
    for (x, y, w, h) in anchor_boxes:
        region = img[y:y+h, x:x+w]  # Crop the region of the image
        region_resized = cv2.resize(region, (224, 224)) / 255.0
        region_expanded = np.expand_dims(region_resized, axis=0)
        
        # Predict on the cropped region
        prediction = model.predict(region_expanded)
        predicted_class = np.argmax(prediction)
        
        # Append if disease detected
        detected_boxes.append((x, y, w, h, predicted_class))
    
    return detected_boxes

# Test on an input image
img = cv2.imread('/kaggle/input/sugarcane-img89/imag89.jpg')
detected_diseases = multi_disease_detection(model, img, anchor_boxes)

# Display detected regions and their predicted classes
for (x, y, w, h, disease_class) in detected_diseases:
    print(f"Disease detected in box [{x}, {y}, {w}, {h}] with predicted class: {disease_class}")

    
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image

# Define Anchor Boxes (x, y, width, height)
anchor_boxes = [
    (50, 100, 100, 150),  # Anchor box 1
    (150, 200, 120, 160),  # Anchor box 2
    (250, 300, 140, 180)   # Anchor box 3
]

# Dictionary for class labels
class_labels = ['Banded Chlorosis', 'Brown Spot', 'BrownRust', 'Dried Leaves', 'Grassy shoot', 'Healthy Leaves', 'Pokkah Boeng', 'Sett Rot', 'Viral Disease', 'Yellow Leaf', 'smut']

# Function to perform disease detection and display results on the image
def multi_disease_detection(model, img_path, anchor_boxes, class_labels):
    # Load the image
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for display
    
    # Iterate over each anchor box
    for (x, y, w, h) in anchor_boxes:
        # Crop the region of interest
        region = img[y:y+h, x:x+w]
        region_resized = cv2.resize(region, (224, 224)) / 255.0
        region_expanded = np.expand_dims(region_resized, axis=0)
        
        # Predict the disease class in the cropped region
        prediction = model.predict(region_expanded)
        predicted_class = np.argmax(prediction)
        predicted_label = class_labels[predicted_class]
        
        # Draw the anchor box on the image
        color = (255, 0, 0)  # Red color for the anchor box
        cv2.rectangle(img_rgb, (x, y), (x + w, y + h), color, 2)
        
        # Put the predicted label above the box
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(img_rgb, predicted_label, (x, y - 10), font, 0.5, color, 2)
    
    # Display the result using Matplotlib
    plt.figure(figsize=(8, 8))
    plt.imshow(img_rgb)
    plt.title('Detected Diseases with Anchor Boxes')
    plt.axis('off')  # Hide the axes
    plt.show()

# Path to the input test image
test_img_path = '/kaggle/input/sugarcane-img-dis-9026/sugarcane-disease_9026img/test/Grassy shoot/Grassy-Shoot-Management-in-Sugarcane-1.jpg'

# Call the function with the trained model, test image path, and anchor boxes
multi_disease_detection(model, test_img_path, anchor_boxes, class_labels)


    
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image

# Define Anchor Boxes (x, y, width, height)
anchor_boxes = [
    (50, 100, 100, 150),  # Anchor box 1
    (150, 200, 120, 160),  # Anchor box 2
    (250, 300, 140, 180)   # Anchor box 3
]

# Dictionary for class labels

class_labels = ['Banded Chlorosis', 'Brown Spot', 'BrownRust', 'Dried Leaves', 'Grassy shoot', 'Healthy Leaves', 'Pokkah Boeng', 'Sett Rot', 'Viral Disease', 'Yellow Leaf', 'smut']

# Function to perform disease detection and display results on the image
def multi_disease_detection(model, img_path, anchor_boxes, class_labels):
    # Load the image
    img = cv2.imread(img_path)
    
    if img is None:
        raise ValueError(f"Image at path {img_path} could not be loaded.")
    
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for display

    # Iterate over each anchor box
    for (x, y, w, h) in anchor_boxes:
        # Crop the region of interest
        region = img[y:y+h, x:x+w]
        region_resized = cv2.resize(region, (224, 224)) / 255.0
        region_expanded = np.expand_dims(region_resized, axis=0)
        
        # Predict the disease class in the cropped region
        prediction = model.predict(region_expanded)
        predicted_class = np.argmax(prediction)
        predicted_label = class_labels[predicted_class]
        
        # Draw the anchor box on the image
        color = (255, 0, 0)  # Red color for the anchor box
        cv2.rectangle(img_rgb, (x, y), (x + w, y + h), color, 2)
        
        # Put the predicted label above the box
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(img_rgb, predicted_label, (x, y - 10), font, 0.5, color, 2)
    
    # Display the result using Matplotlib
    plt.figure(figsize=(8, 8))
    plt.imshow(img_rgb)
    plt.title('Detected Diseases with Anchor Boxes')
    plt.axis('off')  # Hide the axes
    plt.show()

# Path to the input test image
test_img_path = '/kaggle/input/sugarcane-img-dis-9026/sugarcane-disease_9026img/test/Banded Chlorosis/img0089 (18).png'

# Call the function with the trained model, test image path, and anchor boxes
multi_disease_detection(model, test_img_path, anchor_boxes, class_labels)

    
    

    
# Save the model to an H5 file
model.save('mobilenet_plant_village_model.h5')
    
```
